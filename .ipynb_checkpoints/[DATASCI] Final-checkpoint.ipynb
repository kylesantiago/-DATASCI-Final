```python
import numpy as np 
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
```

```python
intPerYearMRT = pd.read_csv('DATA\interruptionsPerYear.csv')
```

```python
intPerYearMRT.head()
```

```python
intPerYearLRT = pd.read_csv('DATA\interruptionsPerYearLRT.csv')
intPerYearLRT.head()
```

```python
tempDFMRT = intPerYearMRT.drop(intPerYearMRT.columns[[0]], axis=1)

fig = plt.figure(figsize=(15, 15))

ax = fig.add_subplot(1,2,1)
im = ax.imshow(tempDFMRT,cmap=cm.Reds)

ax.set_title("MRT Yearly Interruptions")
ax.set_yticks(np.arange(0.5, len(intPerYearMRT.iloc[:,0]), 1))
ax.set_xticks(np.arange(0.5, len(tempDFMRT.columns), 1))

ax.set_xticklabels(tempDFMRT.columns)
ax.set_yticklabels(intPerYearMRT.iloc[:,0])

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(intPerYearMRT.iloc[:,0])):
    for j in range(len(tempDFMRT.columns)):
        text = ax.text(j, i, tempDFMRT.iloc[i,j],
                       ha="center", va="center", color="black")

tempDFLRT = intPerYearLRT.drop(intPerYearLRT.columns[[0]], axis=1)
        
ax = fig.add_subplot(1,2,2)
im = ax.imshow(tempDFLRT,cmap=cm.Reds)

ax.set_title("LRT Yearly Interruptions")
ax.set_yticks(np.arange(0.5, len(intPerYearLRT.iloc[:,0]), 1))
ax.set_xticks(np.arange(0.5, len(tempDFLRT.columns), 1))

ax.set_xticklabels(tempDFLRT.columns)
ax.set_yticklabels(intPerYearLRT.iloc[:,0])

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(intPerYearLRT.iloc[:,0])):
    for j in range(len(tempDFLRT.columns)):
        text = ax.text(j, i, tempDFLRT.iloc[i,j],
                       ha="center", va="center", color="black")

fig.tight_layout()
```

```python
fig = plt.figure(figsize=(20, 5))

ax = fig.add_subplot(1,2,1)
ax.plot(tempDFLRT.sum(axis = 0, skipna = True))
ax.set_ylabel('Number of Interruptions')
ax.set_xlabel('Years')
ax.set_title('LRT Interruptions')


ax = fig.add_subplot(1,2,2)
ax.plot(tempDFMRT.sum(axis = 0, skipna = True))
ax.set_ylabel('Number of Interruptions')
ax.set_xlabel('Years')
ax.set_title('MRT Interruptions')
plt.xticks(rotation=45)
```
